```python
import pandas as pd
import numpy as np
```

Guruhlash. .groupby()
.groupby() meotdi yordamida jadvaldagi ma'lumotlarni biror ustun qiymatlari guruhlarga ajratib tahlil qilishimiz mumkin.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

Yuqoridagi jadvalda turli avtomobil brandlari haqida ma'lumotlar jamlangan. Keling boshlanishiga har bir brand uchun ustunlarning o'rtacha qiymatini topamiz:

```python
df.groupby('company').mean()
```

Yuqoridagi natijani tahlil qilsak, alfa-romero mashinalarning o'rtacha narxi (price) 15498$, honda mashinalarining o'rtacha ot kuchi (horsepower) 92.3 ga teng ekan.

Bitta ustun uchun qiymatlarni olish ham mumkin:



```python
df.groupby('company').price.max()
```

Guruhlashda bir emas, bir nechta ustunlarni ham ko'rsatish mumkin:

```python
df.groupby(['company','body-style']).mean()
```

Jadvaldan xulosa qilsak, volvo firmasinig sedan turidagi mashinalarining o'rtacha qiymati 12940
13415, wagon turidagi mashinalarning o′rtacha narxi esa 13415.

.groupby() yordamida nafaqat o'rta balki boshqa qiymatlarni ham hisoblash mumkin:

```python
df.groupby('company').max()
```

.agg() metodi yordamida esa birdaniga bir nechta statistik ma'lumotlari olishimiz ham mumkin:

```python
df.groupby('company').agg([len,min,max])  #len - guruhdagi elementlar soni, min - guruh uchun minimum, max - guruh uchun maximum
```

.agg() yordamida istalgan funksiyalarni chaqirishimiz mumkin, misol uchun numpy dagi mean (o'rta qiymat) funiksiyasini ham qo'shamiz. Natijalarni faqat price ustuni uchun chiqaramiz:

```python
df.groupby('body-style').price.agg([len,min,max,np.mean])
```

```python
df.groupby('body-style').price.agg([len,min,max,lambda x: np.median(x)])
```
